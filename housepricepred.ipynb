```python
def convert(x):
    temp=x.split('-')
    if len(temp)==2:
        return(float(temp[0])+float(temp[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df['total_sqft']=df['total_sqft'].apply(convert)
df.head()
```

```python
df['bhk']=df['size'].str.extract('(\d+)').astype(int)
df[df['bhk'] > 20]
```

```python
df['pric_per_sqf']=df['price']*1000000 / df['total_sqft']
```

```python
df['location']=df['location'].apply(lambda x: x.strip())
location_count= df['location'].value_counts()
```

```python
location_count_less_10=location_count[location_count <= 10]
print(location_count_less_10)
```

```python
df['location']=df['location'].apply(lambda x: 'other' if x in location_count_less_10 else x)
```

```python
df.pric_per_sqf.describe()
```

```python
def remove_outlier(dt):
    dt_out=pd.DataFrame()
    for key,subdt in dt.groupby('location'):
        m=np.mean(subdt.pric_per_sqf)
        st=np.std(subdt.pric_per_sqf)
        gen_dt=subdt[(subdt.pric_per_sqf >(m-st))&(subdt.pric_per_sqf <= (m+st))]
        dt_out= pd.concat([dt_out,gen_dt],ignore_index=True)
    return dt_out
df=remove_outlier(df)
```

```python
df.to_csv("cleaned_data.csv")
x=df.drop(columns=['price'])
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge 
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score

```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
column_trans=make_column_transformer((OneHotEncoder(sparse_output=False),['location']),remainder='passthrough')
```

```python
scaler=StandardScaler()
lr=LinearRegression()
pipe=make_pipeline(column_trans,scaler,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_lr=pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred_lr)
```

```python
lasso=Lasso()
pipe=make_pipeline(column_trans,scaler,lasso)
pipe.fit(x_train,y_train)
```

```python
y_pred_lasso=pipe.predict(x_test)
r2_score(y_test,y_pred_lasso)
```

```python
ridge=Ridge()
pipe=make_pipeline(column_trans,scaler,ridge)
pipe.fit(x_train,y_train)
```

```python
y_pred_ridge=pipe.predict(x_test)
r2_score(y_test,y_pred_ridge)
```

```python
print("No Regularization: ", r2_score(y_test, y_pred_lr))
print("Lasso: ", r2_score(y_test, y_pred_lasso))
print("Ridge: ", r2_score(y_test, y_pred_ridge))
```
